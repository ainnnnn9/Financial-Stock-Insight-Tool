```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 选一个股票（先用AAPL测试）
ticker = "AAPL"

# 获取数据
stock = yf.Ticker(ticker)
hist = stock.history(period="1y")

# 检查数据
print(hist.head())

# 画图
plt.figure(figsize=(10,5))
plt.plot(hist.index, hist['Close'])
plt.title(f"{ticker} Stock Price (1 Year)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
# 计算简单指标
avg_price = hist['Close'].mean()
volatility = hist['Close'].std()

print("Average Price:", round(avg_price,2))
print("Volatility:", round(volatility,2))
```
